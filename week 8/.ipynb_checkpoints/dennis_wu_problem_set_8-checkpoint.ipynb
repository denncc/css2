## Block 1

```python
# Q1.Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

## Block 2

```python
# Q2. Build the function
def main(X, y, with_plot = False):
    # for extra credit, determine whether the feature is age or sex, which 
    # accordingly fillna or dummy code the feature
    if X.columns[0] == "age":
        # fillna the feature with the average value of the column
        X.age = X.age.fillna(X.age.mean())
    elif X.columns[0] == "sex":
        X = X.replace({"male": 1, "female": 0})
    # Use a train and test split with a random_seed = 1
    Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,random_state=1)
    # Use decision tree binary classification
    tree_model = tree.DecisionTreeClassifier(max_depth=2)
    # traing the model
    tree_mode = tree_model.fit(Xtrain, ytrain)
    # diff of valid and test score
    validation_score = tree_model.score(Xtrain, ytrain)
    test_score = tree_model.score(Xtest, ytest)
    diff = validation_score - test_score
    print(f"Validation score is {validation_score}, and test score is {test_score}")
    print("Difference between testing and training validation scores is", diff)
    if diff > 0.1:
        print("The difference between testing and training validation scores is greater than 0.1. hence overfitting.")
    else:
        print("The model doesn't overfit.")
    if with_plot:
        # initialize subplots
        fig,axes = plt.subplots(2,1,figsize=[8, 10])
        # plot boxplot on first subplot
        sns.boxplot(y, X[X.columns[0]], ax = axes[0])
        # plot ROC curve on second subplot
        fpr_tree,tpr_tree,thresholds_tree = metrics.roc_curve(ytest,tree_model.predict_proba(Xtest)[:,1])
        axes[1].plot(fpr_tree, tpr_tree)
        # set labels
        axes[0].set_title(f"Boxplot of {X.columns[0]} Depending on Survival")
        axes[0].set_xlabel(y.name)
        axes[0].set_ylabel(f'Distribution of {X.columns[0]}')
        axes[1].set_title("ROC Curve")
        axes[1].set_xlabel("false positive rate")
        axes[1].set_ylabel('true positive rate')
        # plot ground truth of ROC
        axes[1].plot([0,1], [0,1], color = [0,0,0], linestyle = ":")
    return
```

## Block 3

```python
# Q3. Load the data using sns.load_dataset('titanic')
df = sns.load_dataset('titanic')
df
```

## Block 4: Apply the model on pclass and survived

```python
# Q4. Apply the model on pclass and survived
y = df.survived
X = df[["pclass"]]
main(X, y, with_plot = True)
```

## Block 5: Apply the model on fare and survived

```python
# Q5. Apply the model on fare and survived
y = df.survived
X = df[["fare"]]
main(X, y, with_plot = True)
```

According to the results of the decision tree models run on pclass and fare, it's evident to find out that most of the people survived tend be more at higher pclass(closer to 1) and spend more money on the fare. Therefore, I strongly recommend Pat to do so. Moreover, according to the validation and test score, although fare has a slightly higher validation score, its test score is smaller than that of the pclass, also leading to fare's overfitting on the data. Therefore, in terms of the model performance, pclass is a better predictor of survival between the two columns.
