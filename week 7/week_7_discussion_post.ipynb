```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
temperature = pd.read_csv("HadCRUT4-gl.dat.txt", delim_whitespace=True).transpose().reset_index()
temperature.index = ["Year", "January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December", "Annual"]
temperature = temperature.transpose()
```

```python
yearly = temperature["Annual"].dropna()
yearly.index = pd.Series(range(1850, 2022, 1))
yearly = yearly.astype(float)
fig, axis = plt.figure()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,2,1.3])
sns.lineplot(data = yearly, ax = axes)
axes.set(xlabel='Year', ylabel='Temperature Anomalies')
axes.set_title("Global Annual Average Temperature Anomalies from 1850 - 2022")
axes.plot([1850, 2021], [0, 0], lw=2)
```
