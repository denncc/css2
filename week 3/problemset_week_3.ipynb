## Importing the libraries

```python
import pandas as pd
import numpy as np
```

## Web Scraping the data

```python
table1 = pd.read_html("https://en.wikipedia.org/wiki/MonsterVerse")[3]
table2 = pd.read_html("https://en.wikipedia.org/wiki/List_of_Fast_%26_Furious_characters")[2]
```

## The function that cleans the data

```python
def clean_table(table):
    res = table.copy()
    res.columns = [col[1] for col in res.columns]
    res = res.set_index(res.columns[0])
    res = res.loc[~res.eq(res.iloc[:, 0], axis=0).all(1)]
    return res
```

## Running the function on table1

```python
res1 = clean_table(table1)
res1
```

## Export the csv result of table1

```python
res1.to_csv("res1.csv", sep = "\t")
```

## Running the function on table2

```python
res2 = clean_table(table2)
res2
```

## Export the csv result of table2

```python
res2.to_csv("res2.csv", sep = "\t")
```
