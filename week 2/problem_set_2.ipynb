```python
import seaborn as sns
import matplotlib.pyplot as plt

anscombe = sns.load_dataset("anscombe")
anscombe
```

```python
fig, axes = plt.subplots(2,2, figsize = (10, 10))
data_letter = ["I", "II", "III", "IV"]
x_labels = ["x1", "x2", "x3", "x4"]
y_labels = ["y1", "y2", "y3", "y4"]
for i in range(4):
    axis = axes.flatten()[i]
    axis.set_xlim([2,20])
    axis.set_ylim([2,14])
    subdata = anscombe[anscombe["dataset"] == data_letter[i]]
#     axis.plot(subdata["x"], subdata["y"],
#              linestyle = "",
#              marker = ".")
    sns.regplot(x = "x", y = "y", data = subdata, ax = axis)
#     g = sns.regplot("x","y",
#                 data = anscombe[anscombe["dataset"] == data_letter[i]],
#                 ax = axis,
#                 truncate = False)
#     g.set(ylim = (2,14))
#     g.set(xlim = (2,20))
    axis.set_title("Dataset " + data_letter[i])
    axis.text(15, 4, "x mean: " + str(subdata["x"].mean())[:5])
    axis.text(11, 3, "x standard deviation: " + str(subdata["x"].std())[:5])
    axis.text(6.3,13, "y mean:" + str(subdata["y"].mean())[:5])
    axis.text(2.3, 12, "y standard deviation: " + str(subdata["y"].std())[:5])
plt.tight_layout()
```
