```python
# 1. Load relevant libraries
import pandas as pd
import numpy as np
import seaborn as sns
import openpyxl
```

```python
# Load the data, show the shape of the data AND the head of the data
iris = pd.read_excel("messed_up_iris.xlsx", engine = "openpyxl", index_col = 0)
print(iris.shape)
display(iris.head())
```

```python
# Remove columns where there are more than 50% of the data missing.
iris = iris.loc[:,(~((iris.isna().sum(axis = 0) / iris.shape[0]) > 0.5)).values]
# Remove rows where there are more than 50% of the data missing.
iris = iris.loc[(~((iris.isna().sum(axis = 1) / iris.shape[1]) > 0.5)).values,:]
print(iris.shape)
```

```python
# correct the typos
iris.species = np.where(iris.species == "virginia", "virginica", iris.species)
iris.species = np.where(iris.species == "west virginia", "virginica", iris.species)
iris.species = np.where(iris.species == "versicolr", "versicolor", iris.species)
iris.species = np.where(iris.species == "seotsa", "versicolor", iris.species)

iris.origin = np.where(iris.origin == "uas", "usa", iris.origin)
iris.origin = np.where(iris.origin == "euarope", "europe", iris.origin)
# Dummy code the categorical data
iris = iris.join(pd.get_dummies(iris.species, drop_first = True))
iris = iris.join(pd.get_dummies(iris.origin, drop_first = True))
iris.head()
```

```python
# Drop the redundant columsn from the dataframe
species = iris.species
iris = iris.drop(columns = ["species", "origin"])
iris.head()
```

```python
# Remove the outliers above 2 standard deviations
iris.sepal_length = np.where(
    iris.sepal_length > iris.sepal_length.mean() + 2 * iris.sepal_length.std(), np.nan, iris.sepal_length
)
iris.sepal_length = np.where(
    iris.sepal_length < iris.sepal_length.mean() - 2 * iris.sepal_length.std(), np.nan, iris.sepal_length
)
iris.sepal_width = np.where(
    iris.sepal_width > iris.sepal_width.mean() + 2 * iris.sepal_width.std(), np.nan, iris.sepal_width
)
iris.sepal_width = np.where(
    iris.sepal_length < iris.sepal_width.mean() - 2 * iris.sepal_width.std(), np.nan, iris.sepal_width
)
iris.petal_length = np.where(
    iris.petal_length > iris.petal_length.mean() + 2 * iris.petal_length.std(), np.nan, iris.petal_length
)
iris.petal_length = np.where(
    iris.petal_length < iris.petal_length.mean() - 2 * iris.petal_length.std(), np.nan, iris.petal_length
)
iris.petal_width = np.where(
    iris.petal_width > iris.petal_width.mean() + 2 * iris.petal_width.std(), np.nan, iris.petal_width
)
iris.petal_width = np.where(
    iris.petal_width < iris.petal_width.mean() - 2 * iris.petal_width.std(), np.nan, iris.petal_width
)
```

```python
# Bonus point: replace the missing values from last cell to the median value of the species
med = iris.groupby(["versicolor", "virginica"]).median().drop(columns = ["japan", "usa"])
# fill np.nan of setosa
iris.sepal_length = np.where(iris.sepal_length.isna() & (iris.versicolor == 0) & (iris.virginica == 0),
                            med["sepal_length"][(0,0)], iris.sepal_length)
iris.sepal_width = np.where(iris.sepal_width.isna() & (iris.versicolor == 0) & (iris.virginica == 0),
                            med["sepal_width"][(0,0)], iris.sepal_width)
iris.petal_length = np.where(iris.petal_length.isna() & (iris.versicolor == 0) & (iris.virginica == 0),
                            med["petal_length"][(0,0)], iris.petal_length)
iris.petal_width = np.where(iris.petal_width.isna() & (iris.versicolor == 0) & (iris.virginica == 0),
                            med["petal_width"][(0,0)], iris.petal_width)
# fill np.nan of versicolor
iris.sepal_length = np.where(iris.sepal_length.isna() & (iris.versicolor == 1) & (iris.virginica == 0),
                            med["sepal_length"][(1,0)], iris.sepal_length)
iris.sepal_width = np.where(iris.sepal_width.isna() & (iris.versicolor == 1) & (iris.virginica == 0),
                            med["sepal_width"][(1,0)], iris.sepal_width)
iris.petal_length = np.where(iris.petal_length.isna() & (iris.versicolor == 1) & (iris.virginica == 0),
                            med["petal_length"][(1,0)], iris.petal_length)
iris.petal_width = np.where(iris.petal_width.isna() & (iris.versicolor == 1) & (iris.virginica == 0),
                            med["petal_width"][(1,0)], iris.petal_width)
# fill np.nan of vriginica
iris.sepal_length = np.where(iris.sepal_length.isna() & (iris.versicolor == 0) & (iris.virginica == 1),
                            med["sepal_length"][(0,1)], iris.sepal_length)
iris.sepal_width = np.where(iris.sepal_width.isna() & (iris.versicolor == 0) & (iris.virginica == 1),
                            med["sepal_width"][(0,1)], iris.sepal_width)
iris.petal_length = np.where(iris.petal_length.isna() & (iris.versicolor == 0) & (iris.virginica == 1),
                            med["petal_length"][(0,1)], iris.petal_length)
iris.petal_width = np.where(iris.petal_width.isna() & (iris.versicolor == 0) & (iris.virginica == 1),
                            med["petal_width"][(0,1)], iris.petal_width)
```

```python
# Bonus: bin the petal width column into low, medium, and high
iris["petal_width_bins"] = pd.cut(iris.petal_width, bins = [0, 0.9, 1.8, 2.7], labels = [1,2,3])
iris.head()
```

```python
# Show there are no missing values
print(iris.isna().sum())
# Use describe method to show off the final dataset
display(iris.describe())
```
